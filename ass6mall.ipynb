```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering  # For Hierarchical Clustering
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Mall_Customers.csv')

# Check for missing values
print(data.isnull().sum())

# Label Encoding for 'Genre' column (Gender: Male = 0, Female = 1)
le = LabelEncoder()
data['Genre'] = le.fit_transform(data['Genre'])

# Data Transformation: Standardizing numerical features for clustering
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data.iloc[:, 3:])

# KMeans Clustering
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(data_scaled)

# Plotting KMeans Clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Annual Income (k$)'], y=data['Spending Score (1-100)'], hue=kmeans_labels, palette='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, color='red', label='Centroids')
plt.title('KMeans Clustering: Annual Income vs Spending Score')
plt.legend()
plt.show()

# Hierarchical Clustering (Agglomerative)
agg_clust = AgglomerativeClustering(n_clusters=5)
agg_labels = agg_clust.fit_predict(data_scaled)

# Plotting Agglomerative Clustering Results
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Annual Income (k$)'], y=data['Spending Score (1-100)'], hue=agg_labels, palette='coolwarm')
plt.title('Agglomerative Clustering: Annual Income vs Spending Score')
plt.legend()
plt.show()

# Optional: Comparing the two methods in terms of centroid positions or silhouette score
from sklearn.metrics import silhouette_score
print(f'Silhouette Score for KMeans: {silhouette_score(data_scaled, kmeans_labels)}')
print(f'Silhouette Score for Agglomerative: {silhouette_score(data_scaled, agg_labels)}')

# If you still want to split the data (for example, for validation or testing):
from sklearn.model_selection import train_test_split
x_train, x_test = train_test_split(data_scaled, test_size=0.3, random_state=42)

```
